```python
import numpy as np
import pandas as pd
import json
from operator import itemgetter
from itertools import chain
from collections import OrderedDict
import seaborn as sb
import matplotlib.pyplot as plt
import ast
#from wordcloud import WordCloud, STOPWORDS
csv_data = pd.read_excel('alpha_occur_cleaned_data.xlsx', header = None)
```

```python
genredict = {}

def gettags(index):

    taglist = []
    taglist = ast.literal_eval(csv_data[3][index])
    #print(taglist)
    return taglist

def getgenre(index):

    genrelist = []
    genrelist = ast.literal_eval(csv_data[2][index])
    #print(genrelist)
    return genrelist

def addtagtogenre(genre, taglist):
    if genre not in genredict:
        genredict[genre] = {}
    for tag in taglist:
        if tag in genredict[genre]:
            genredict[genre][tag] += 1
        else:
            genredict[genre][tag] = 1

def sortdict(dict):
    sorted_dict = OrderedDict(sorted(dict.items(), key=lambda x: x[1], reverse=True))
    return sorted_dict


for i in range(1, len(csv_data), 1):
    genrelist = getgenre(i)
    taglist = gettags(i)
    for genre in genrelist:
        addtagtogenre(genre, taglist) 

```

# Genre finding count and Graph

```python
df1 = pd.read_excel('alpha_occur_cleaned_data.xlsx')
```

```python
df1.head()
```

```python
processing_data = df1[['title', 'id', 'genres', 'keywords']]
```

```python
processing_data.isna().sum()
```

```python
processing_data.head()
```

```python
import ast
```

```python
# For loop on all the strings
for i in range(processing_data.shape[0]):
    processing_data['genres'][i] = ast.literal_eval(processing_data['genres'][i])
```

```python
# Selecting only those rows which have an actual genre
genre_present = processing_data['genres'] != '[]'

# Series of the genres present in the movies_metadata
genres = processing_data['genres'][genre_present]
```

```python
from sklearn.preprocessing import MultiLabelBinarizer
mlb = MultiLabelBinarizer()

labels = mlb.fit_transform(genres)
label_classes = mlb.classes_
```

```python
label_classes
```

```python
label_data = pd.DataFrame(labels, columns=label_classes)
```

```python
val = {}
for x in label_classes :
    val.update({x:label_data[x].value_counts()[1]})
```

```python
sorted_val = sorted(val.items(), key=lambda kv: kv[1], reverse=True)
```

```python
val_pd = pd.DataFrame.from_dict(sorted_val, orient='columns')
val_pd.rename(columns={0: "Genre", 1: "Count"}, inplace = True) 
```

```python
val_pd
```

```python
val_pd['Genre'] = val_pd['Genre'].astype('category')
```

```python
val_pd.dtypes
```

```python
sb.catplot(y = 'Genre',x = 'Count', data = val_pd,kind = "bar", height = 5)
```
